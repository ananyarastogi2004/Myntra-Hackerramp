```python
# Function to categorize subcategory based on keywords in name
def categorize_subcategory(subcategory):
    if 'cargo' in subcategory.lower():
        return 'Cargo'
    elif 'jogger' in subcategory.lower():
        return 'Jogger'
    elif 'pants' in subcategory.lower():
        return 'Pants'
    elif 'trousers' in subcategory.lower():
        return 'Trousers'
    elif 'denim jacket' in subcategory.lower():
        return 'Denim Jacket'
    elif 'bomber jacket' in subcategory.lower():
        return 'Bomber Jacket'
    elif 'cropped jacket' in subcategory.lower():  # Removed extra space
        return 'Cropped Jacket'
    elif 'padded' in subcategory.lower():
        return 'Padded Jacket'
    elif 'puffer' in subcategory.lower():
        return 'Puffer Jacket'

# Create a mask for the specified categories
jeans_mask = df['category'].isin(['Trousers & Capris', 'Jackets & Coats'])

# Apply the function to update the 'subcategory' column
df.loc[jeans_mask, 'subcategory'] = df.loc[jeans_mask, 'subcategory'].apply(categorize_subcategory)
```

```python
# Function to categorize subcategory based on keywords in name
def categorize_subcategory(name):
    if 'a-line' in name.lower():
        return 'A-Line Skirt'
    elif 'pencil' in name.lower():
        return 'Pencil Skirt'
    elif 'flared' in name.lower():
        return 'Flared Skirt'
    elif 'ruffled' in name.lower():
        return 'Ruffled Skirt'
    elif 'wrap' in name.lower():
        return 'Wrap Skirt'
    elif 'straight' in name.lower():
        return 'Straight Skirt'
    elif 'mini' in name.lower():
        return 'Mini Skirt'
    elif 'shorts' in name.lower():
        return 'Shorts'

# Apply the function to create a new 'subcategory' column with high rise or mid rise categories
jeans_mask = df['category'] == 'Shorts & Skirts'
df.loc[jeans_mask, 'subcategory'] = df.loc[jeans_mask, 'name'].apply(categorize_subcategory)
```

```python

import ast  # For literal_eval

def safe_literal_eval(x):
    try:
        return ast.literal_eval(x)
    except (ValueError, SyntaxError):
        return x  # Return the original string if parsing fails

# Convert the 'p_attributes' column from string to dictionary, handling potential errors
df['p_attributes'] = df['p_attributes'].apply(safe_literal_eval)

# Extract 'fabric' and 'occasion' from the dictionary
df['fabric'] = df['p_attributes'].apply(lambda x: x.get('Fabric') if isinstance(x, dict) else None)
df['occasion'] = df['p_attributes'].apply(lambda x: x.get('Occasion') if isinstance(x, dict) else None)

df.head()
```

```python
df['subcategory'] = df['subcategory'].replace('Cropped Top', 'Crop Top')
```

```python
# Function to categorize subcategory based on keywords in name
def categorize_subcategory(name):
    if 'wide leg high-rise' in name.lower():
        return 'Wide Leg High-Rise'
    elif 'wide leg mid-rise' in name.lower():
        return 'Wide Leg Mid-Rise'
    elif 'skinny fit high-rise' in name.lower():
        return 'Skinny Fit High-Rise'
    elif 'skinny fit mid-rise' in name.lower():
        return 'Skinny Fit Mid-Rise'
    elif 'straight fit high-rise' in name.lower():
        return 'Straight Fit High-Rise'
    elif 'straight fit mid-rise' in name.lower():
        return 'Straight Fit Mid-Rise'
    elif 'flared high-rise' in name.lower():
        return 'Flared High-Rise'
    elif 'flared mid-rise' in name.lower():
        return 'Flared Mid-Rise'
    elif 'bootcut mid-rise' in name.lower():
        return 'Bootcut Mid-Rise'
    elif 'bootcut high-rise' in name.lower():
        return 'Bootcut High-Rise'

# Apply the function to create a new 'subcategory' column with high rise or mid rise categories
jeans_mask = df['category'] == 'Jeans'
df.loc[jeans_mask, 'subcategory'] = df.loc[jeans_mask, 'name'].apply(categorize_subcategory)
```

```python
# Assuming your DataFrame is named 'df'
occasion_to_drop = ['Maternity', 'Ethnic', 'Daily', 'Western']
df = df[~df['occasion'].isin(occasion_to_drop)]
```

```python
valid_subcategories = [
    'Crop Top', 'Wrap Top', 'Peplum Top', 'A-Line Top','Shirt Top', 'Tank Top', 'Vest Top', 'Ribbed Top',
    'Wide Leg High-Rise', 'Wide Leg Mid-Rise', 'Skinny Fit High-Rise','Skinny Fit Mid-Rise', 'Straight Fit High-Rise', 'Straight Fit Mid-Rise','Flared High-Rise', 'Bootcut Mid-Rise',
    'Sweatshirt','Hooded Sweatshirt',
    'Cargo', 'Jogger', 'Pants', 'Trousers',
    'Denim Jacket', 'Bomber Jacket', 'Cropped Jacket', 'Padded Jacket', 'Puffer Jacket',
    'A-Line Skirt', 'Pencil Skirt', 'Flared Skirt', 'Ruffled Skirt', 'Wrap Skirt', 'Straight Skirt', 'Mini Skirt', 'Shorts',
    'Blazer'
]

# Initialize an empty DataFrame to store the mini dataset
mini_dataset = pd.DataFrame()

# Iterate over each subcategory and sample 10 rows
for subcategory in valid_subcategories:

    filtered_rows = df[(df['subcategory'] == subcategory) & (df['category'].isin(['Top', 'Jeans','Sweaters & Sweatshirts','Trousers & Capris','Shorts & Skirts','Jackets & Coats','Blazers & WaistCoats']))]

    # Print the number of rows found for the current subcategory
    print(f"{subcategory}: {len(filtered_rows)} rows found")

    # Sample 10 rows (or fewer if not enough rows available)
    #sampled_rows = filtered_rows.sample(n=10, random_state=1) if len(filtered_rows) >= 10 else filtered_rows

    # Append sampled rows to the mini dataset
    mini_dataset = pd.concat([mini_dataset, filtered_rows])

# Reset index for the mini dataset
mini_dataset = mini_dataset.reset_index(drop=True)

# Display the shape of the mini dataset
print(f"Mini dataset shape: {mini_dataset.shape}")

# Display the mini dataset
print(mini_dataset)
```

```python
mini_dataset.to_csv('mini_dataset.csv', index=False)
```
