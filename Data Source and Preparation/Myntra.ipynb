```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/sample_data/myntra.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df = df.drop(columns=['ratingCount','avg_rating'])
```

```python
df.head()
```

```python
np.shape(df)
```

```python
# List of words to filter out
filter_words = ['saree', 'dupatta', 'kurti','Salwar suit','Sharara','kurta','Palazzos','Lehanga']

# Create a regex pattern to match any of the filter words
pattern = '|'.join(filter_words)

# Drop rows where 'products' column contains any of the filter words
df1 = df[~df['products'].str.contains(pattern, case=False)]
df1.head()
```

```python
np.shape(df1)
```

```python
# Finding unique values in the 'products' column
unique_products = df1['products'].unique()

# Displaying all unique products
pd.set_option('display.max_rows', 1000)
print(unique_products)
```

```python
# List of words to keep
keep_words = ['top', 'jeans', 'skirt', 'pant', 'trousers', 'denim', 'jacket','blazer','shirt','sweatshirt','T-Shirt']

# Create a regex pattern to match any of the keep words
pattern = '|'.join(keep_words)

# Keep rows where 'products' column contains any of the keep words
df2 = df1[df1['products'].str.contains(pattern, case=False)]
df2.head()
```

```python
# Get the count of unique products
unique_products_count = df2['products'].nunique()

# Display the count of unique products
print(f"Count of unique products: {unique_products_count}")
```

```python
np.shape(df2)
```

```python
# Function to determine category based on keywords in the products column
def determine_category(product):
    if 'top' in product.lower():
        return 'Top'
    elif 't-shirt' in product.lower():
        return 'T-Shirt'
    elif 'shirt' in product.lower():
        return 'Shirt'
    elif 'jeans' in product.lower():
        return 'Jeans'
    elif 'skirt' in product.lower():
        return 'Shorts & Skirts'
    elif 'shorts' in product.lower():
        return 'Shorts & Skirts'
    elif any(word in product.lower() for word in ['trousers', 'pants', 'cargo', 'joggers']):
        return 'Trousers & Capris'
    elif 'jacket' in product.lower():
        return 'Jackets & Coats'
    elif 'blazer' in product.lower():
        return 'Blazers & WaistCoats'
    elif 'sweatshirt' in product.lower():
        return 'Sweatshirts & Sweaters'

# Create a new column 'category'
df2['category'] = df2['products'].apply(determine_category)
```

```python
df2.head()
```

```python
df2['subcategory'] = df2['products']
df2.head()
```

```python
# Finding unique values in the 'products' column
unique_products = df2['subcategory'].unique()

print(unique_products)
```

```python
df2.to_csv('new_myntra.csv', index=False)
```
